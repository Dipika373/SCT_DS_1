```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px # import the plotly.express module

df = pd.read_csv('/content/drive/MyDrive/skills/population.csv')
df['Males'] = pd.to_numeric(df['Males'].str.replace(',', '')) # remove commas before conversion
df['Females'] = pd.to_numeric(df['Females'].str.replace(',', ''))

pieChart = px.pie(df, values='Females', names='Males', title='Distribution of gender in population') # Pass df to the function
pieChart.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Males'], df['Females'])
plt.xlabel('Males')
plt.ylabel('Females')
plt.title('Population of Males vs Females')
plt.legend()
plt.show()
```

```python
plt.hist2d(df['Males'], df['Females'])
plt.xlabel('Males')
plt.ylabel('Females')
plt.title('Population of Males vs Females')
plt.colorbar()
plt.show()

```

```python
import matplotlib.pyplot as plt
plt.hist(df['Males'], bins=10)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution')
plt.show()
```
